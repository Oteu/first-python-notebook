Who contributes the most to Prop 67?

```python
import pandas as pd
```

Download data source

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

Find column headers

```python
props.info()
```

```python
contribs.info()
```

Get full prop name

```python
props.prop_name.value_counts()
```

Filter dataframe for Prop 67

```python
prop67 = props[props.prop_name == "PROPOSITION 067- REFERENDUM TO OVERTURN BAN ON SINGLE-USE PLASTIC BAGS."]
```

```python
prop67.head()
```

Merge dataframes prop67 and contribs

```python
merged = pd.merge(prop67, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.prop_name.value_counts().reset_index()
```

Find totals

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
support.amount.sum()
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
oppose.amount.sum()
```

Sort by amount

```python
support.sort_values('amount', ascending=False).head()
```

```python
oppose.sort_values('amount', ascending=False).head()
```

Find total froms supporters and opposition

```python
merged.groupby("committee_position").amount.sum().reset_index().sort_values('amount', ascending=False)
```

```python
support.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values('amount', ascending=False).head()
```

```python
oppose.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values('amount', ascending=False)
```

```python
oppose.contributor_lastname.value_counts()
```

```python
support.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values('amount', ascending=False).head()
```
